```python
import pandas as pd
import os
#from urllib.request import urlretrieve
import numpy as np
import glob
```

```python
lower_secondary = 'Lower Secondary.csv'
df_1 = pd.read_csv(lower_secondary)
print(df_1.head())
```

```python
primary = 'Primary.csv'
df_2 = pd.read_csv(primary)
print(df_2.head())
```

```python
upper_secondary = 'Upper Secondary.csv'
df_3 = pd.read_csv(upper_secondary)
print(df_3.head())
```

```python
frames = [df_1, df_2, df_3]
super_data_frame = pd.concat(frames, axis=0, keys=['a', 'b', 'c'])
pd.set_option('display.max_rows', 606)
print(super_data_frame)
```

```python
path = 'input'
dw_csv_files = glob.glob(path + '/*.csv')

df_list = []

for f in dw_csv_files:
    temp_df = pd.read_csv(f)
    df_list.append(temp_df)
    print(f'Successfully created dataframe for {f} with shape {temp_df.shape}')
    
df_0 = pd.concat(df_list, axis=0)
print(df_0.shape)
df_0.head()
```
